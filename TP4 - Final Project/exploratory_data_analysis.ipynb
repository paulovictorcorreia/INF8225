# Introduction

Notebook for a brief exploratory data analysis of the EEG data.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
plt.style.use("ggplot")
```

```python
eeg_signals = pd.read_csv("data/eeg_signals.csv", index_col=0)
eeg_signals.info()
```

```python
eeg_signals.head()
```

# Signals For Each Condition

```python
np.random.seed(0)
sample_condition_1 = eeg_signals[eeg_signals["y"] == 1].iloc[0]
sample_condition_2 = eeg_signals[eeg_signals["y"] == 2].iloc[0] 
sample_condition_3 = eeg_signals[eeg_signals["y"] == 3].iloc[0] 
sample_condition_4 = eeg_signals[eeg_signals["y"] == 4].iloc[0]
sample_condition_5 = eeg_signals[eeg_signals["y"] == 5].iloc[0]


gs = GridSpec(3, 2, width_ratios=[1, 1], height_ratios=[1, 1, 2])
plt.figure(figsize=(8, 12))


ax1 = plt.subplot(321)
ax1.plot(np.asarray(sample_condition_5), label="Eyes Closed Signal")
plt.legend(loc="best")

ax1 = plt.subplot(322)
ax1.plot(np.asarray(sample_condition_2), label="Tumor Area Signal")
plt.legend(loc="best")

ax1 = plt.subplot(323)
ax1.plot(np.asarray(sample_condition_3), label="Healthy Area Signal")
plt.legend(loc="best")

ax1 = plt.subplot(324)
ax1.plot(np.asarray(sample_condition_4), label="Eyes Closed Signal")
plt.legend(loc="best")

ax1 = plt.subplot(313)
ax1.plot(np.asarray(sample_condition_1), label="Seizure Signal")
plt.legend(loc="best")



plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.plot(np.asarray(sample_condition_1), label="Seizure Signal")
plt.plot(np.asarray(sample_condition_2), label="Tumor Area Signal")
plt.plot(np.asarray(sample_condition_3), label="Healthy Area Signal")
plt.plot(np.asarray(sample_condition_4), label="Eyes Closed Signal")
plt.plot(np.asarray(sample_condition_5), label="Eyes Opened Signal")

plt.title("EEG Signals from UCI Dataset")
plt.legend(loc="lower right")
```
